```python
import pandas as pd
```

```python
df = pd.read_csv('case_data.csv')
##Меняем названия стоблцов
df.columns = ['checkId','date', 'inn', 'customerId', 'product', 'brand', 'tag', 'price', 'amount', 'sum', 'cashback']
df.head()
```

```python
innSeries = df.inn

## Преобразуем экспоненциальные значения
innSeries = innSeries.apply(lambda x: float(x.replace(',', '.')))
pd.to_numeric(innSeries)
innSeries = innSeries.apply(lambda x: '{}'.format(int(x)))

## Создаем серию уникальных ИНН
innList = innSeries.unique()
## Преобразуем в список
innList = innList.tolist()
innList
```

```python
len(innList)
```

```python
## Парсинг ИНН из сервиса list-org.com
import requests
from bs4 import BeautifulSoup
##from HTMLParser import HTMLParseError

urlName = ('https://www.list-org.com/search?type=inn&val=')

for innValue in innList:
    url = ('{}{}'.format(urlName, innValue))
    print(url)       
    response = requests.post(url, headers={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:70.0) Gecko/20100101 Firefox/70.0', 'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8', 
'Accept-Language': 'ru-RU,ru;q=0.8,en-US;q=0.5,en;q=0.3', 'Accept-Encoding': 'gzip, deflate'}, cookies={'PHPSESSID':'null'}, data={'type':'inn'},)
    print(response.status_code)
    ## Получаем ответ в виде HTML
    text = response.text
    soup = BeautifulSoup(text)
    companyList = []
    ## Получаем название компании
    try:
        content = soup.find('div', {'class': 'content'})
        companyName = content.find('div', {'class': 'org_list'}).find('a').get('href')
        company = content.find('div', {'class': 'org_list'}).find('a').text
        print(company)
        ## Создаем серию компаний с индексами-ИНН
        companyList.append(company)
    except Exception:
        companyList.append('NaN')
```

```python
companyInn = pd.Series(data=companyList, index=innList)
companyInn
```
